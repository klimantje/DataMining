# Data exploration

## Loading the data

```python
#libraries for exploration
import pandas as pd
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler, Imputer, normalize
from scipy import stats

# plotting missing values
import missingno as msno

#plotting inline matplotlib
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import model_selection
from sklearn.model_selection import train_test_split, cross_val_score
```

We load the data as a pandas dataframe and exampine the different features.

```python
titanic = pd.read_csv("train_titanic.csv")
```

```python
titanic.head(20)
```

We check whether the attributes have the correct types

```python
titanic.info()
```

## Exploring the variables

We perform some basic data analysis on the different variables. Let's first look at the statistics of the numeric variables. 

```python
titanic.describe()
```

We see that there are records with Fare 0, we want to treat them as missing values so we replace them with NaN so that we can impute them later.

```python
#replace zeroes with nan
titanic.Fare = titanic.Fare.map(lambda x: np.nan if x==0 else x)
```

For the categorical variables we see that people shared a cabin and the most people embarked in Southhampton.

```python
titanic.describe(include=['O'])
```

We check if the types of the fields are right

```python
titanic.dtypes
```

We make a new feature family size and drop name and ticket columns as they are not useful for the models.

```python
titanic["FamilySize"] = titanic["SibSp"] + titanic["Parch"]
# drop useless columns
titanic.drop('Name', axis=1, inplace=True)
titanic.drop('Cabin', axis=1, inplace=True)
titanic.drop('Ticket', axis=1, inplace=True)
```

```python
numeric_features = ["Age", "Fare"]
ordinal_features = ["Pclass", "SibSp", "Parch", "FamilySize"]
nominal_features = ["Sex" , "Embarked"]
target = "Survived"
```

```python
titanic.head()
```

Split the variables in numberic and categorical. 

### Numeric variables

We make histograms and boxplots of the different numerical columns. We see that Fare is very skewed, which is normal, many people bought a cheap ticket and very few bought an expensive one.

```python
### Plotting Numeric Features
# Looping through and Plotting Numeric features


for column in numeric_features:   
    
    # Figure initiation
    fig = plt.figure(figsize=(18,12))
    ### Distribution plot
    sns.distplot(titanic[column].dropna(), ax=plt.subplot(221));
    # X-axis Label
    plt.xlabel(column, fontsize=14);
    # Y-axis Label
    plt.ylabel('Density', fontsize=14);
    ### Boxplot
    sns.boxplot(titanic[column].dropna(), ax=plt.subplot(222));
    # X-axis Label
    plt.xlabel(column, fontsize=14);
    # Y-axis Label
    plt.ylabel("Box", fontsize=14)
    # Adding Super Title (One for a whole figure)
    plt.suptitle('Plots for ' + column, fontsize=18);
 
# Printing Chart
plt.show()

    
```

### Categorical variables

We make a simple countplot for each variable to see how the passengers are distributed over the different categories. We see for instance that more males than females where on board and many people travelled alone.

```python
### Plotting Numeric Features
# Looping through and Plotting Numeric features


for column in nominal_features + ordinal_features:   
    
    # Figure initiation
    fig = plt.figure(figsize=(10,6))
    ### Distribution plot
    sns.countplot(titanic[column].dropna());
    # X-axis Label
    plt.xlabel(column, fontsize=14);
   # Adding Super Title (One for a whole figure)
    plt.suptitle('Count class size for ' + column, fontsize=18);
 
# Printing Chart
plt.show()
```

## Examing the missing values

We use the missing_no package to visualise the missing values. We see that therer are only missing values in Age, Cabin and Embarked.

```python
msno.matrix(titanic)
plt.show()
```

```python
titanic.isnull().sum().sort_values()
```

Only age, cabin, fare  and embarked have missing values.

## Handling missing values

There are a lot of ages and cabins missing. We decide to impute age by the mean and cabin by "Unknown".

Age is not too skewed so we impute by the mean.

```python
sns.distplot(titanic.Age.dropna())
plt.show()
```

```python
titanic.Age= titanic.Age.fillna(titanic.Age.mean())
```

We see that all the missing Fares are Embarked at S. It seems reasonable that fare is dependent on where you embark and the class. 

```python
titanic[np.isnan(titanic.Fare)]
```

So we plot the fares for embarkment in Southampton per class. As it 's fairly skewed we decide to use median instead of mean.

```python
facet = sns.FacetGrid(titanic[~np.isnan(titanic.Fare) & (titanic.Embarked =='S')] , aspect=4, hue='Pclass')
facet.map(sns.kdeplot,'Fare',shade= True)
facet.set(xlim=(0, 80))


fare_median = titanic[(titanic['Embarked'] == 'S')].groupby(["Pclass"])['Fare'].median()
for x in fare_median:
    plt.axvline(x, ls='--', color ='r')
plt.title("Fares per class")
plt.show()
```

So we impute the Fare based on these medians:

```python
titanic["Fare"] = titanic.groupby("Pclass")["Fare"].transform(lambda x: x.fillna(x.median()))
```

There are two passengers where the embarked is missing. Like we did for fare we can try to derive where they embarked from their fare and class. Both passengers have the same class and the same Fare.

```python
titanic[titanic.Embarked.isnull()]
```

We visualize fares per class and embarkment. We see that the most likely embarkment for the missing values is Charlottesville so we impute with "C".

```python
sns.factorplot(x='Embarked',y='Fare', hue='Pclass', kind="box",order=['C', 'Q', 'S'],data=titanic[~titanic.Embarked.isnull()], size = 7, aspect = 2,)
#limit y axis for 
plt.ylim(0,200)
plt.title("Fares per class and city of embarkment")

# Plot the fare of 80 they paid
plt.axhline(y=80, color='r', ls='--')
plt.show()
```

```python
titanic.Embarked = titanic.Embarked.fillna('C')
```

No missing values are left!

```python
titanic.isnull().sum().sort_values()
```

## Exploring the correlation between variables

We now explore the correlation between the variables to see which ones we should consider for the models. We do this by generating a correlation matrix. We see that class and fare are negatively correlated, as we also noticed when we were imputing the null values.

```python
# GETTING Correllation matrix
corr_mat=titanic.drop([target], axis =1).corr(method='pearson')
plt.figure(figsize=(20,10))
sns.heatmap(corr_mat,vmax=1,square=True,annot=True,cmap='cubehelix')
plt.title("Correlation between non-class variables")
plt.show()
```

Obviously Family size is highly correlated with Parch and Sibsp so we can likely drop those after we examine correlation with the target variable.

# Classification

## Splitting in train and testset

We use one hot encoding to convert the nominal variables.

```python
titanic = pd.get_dummies(titanic, columns=nominal_features)
```

```python
titanic.head()
```

We split the dataset in a test and trainset. Our testset size is 40%.

```python
# create training and testing vars
X_train, X_test, y_train, y_test = train_test_split(titanic.drop([target], axis=1), titanic[target], test_size=0.4)
```

## Different classifiers

We have to center and scale age and fare. We hence setup an scikit-learn pipeline. We make a list of models to try:
- logistic regression
- support vector machines
- Random Forest
- K nearest neighbours

For KNN it seems it depends a lot on the number of neighbours so we try to find the best number. That seems to be around 23,

```python
from sklearn.pipeline import Pipeline
models=dict()
train_error =[]
test_error=[]

#Logistic Regression
models.update({"Logistic regression": LogisticRegression(penalty='l2', C=0.001, random_state=0)})
models.update({"SVM": SVC(kernel='rbf', gamma=1/2 , C=1.0,class_weight='balanced', 
         random_state=0,probability=True)})   #loss='squared_hinge'
models.update({"Random Forest": RandomForestClassifier(n_estimators=1000,random_state=0,n_jobs=-1)})
for i in range(1,40):
    models.update({"KNN" + str(i) : KNeighborsClassifier(n_neighbors=i, p=2, metric='minkowski')})



for label, clf in models.items():
    pipeline = Pipeline([('scaler',StandardScaler()),('clf',clf )])
    scores = cross_val_score(
        estimator=pipeline,
        X=X_train,
        y=y_train,
        cv=10,
        scoring='roc_auc',
    )
    print("ROC AUC: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))



```

## Performance on the testset

We check now the performance on the testset and see that KNN does not perform very well. Random forest seems the most promising so we will tweak that.

```python
from sklearn.metrics import roc_curve, auc
#colors = ['black', 'orange', 'blue', 'green','pink']
#linestyles = [':', '--', '-.', '-','-']
plt.figure(figsize=(20,20))
for label,clf in models.items():
    pipeline = Pipeline([('scaler',StandardScaler()),('clf', clf)])
    y_pred = pipeline.fit(X_train, y_train).predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_true=y_test, y_score=y_pred)
    roc_auc = auc(x=fpr, y=tpr)
    plt.plot(
        fpr,
        tpr,
        label='%s (auc = %0.2f)' % (label, roc_auc),
    )
plt.legend(loc='lower right')
plt.plot(
[0, 1],
[0, 1],
linestyle='--',
color='gray',
linewidth=2,

)
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.grid()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

## Different random forest

```python
from sklearn.model_selection import GridSearchCV

rfc = RandomForestClassifier(n_jobs=-1,  oob_score = True) 
pipeline = Pipeline([('scaler',StandardScaler()),('clf', rfc)])
# Use a grid over parameters of interest
param_grid = {"clf__n_estimators":[1000],
    "clf__max_depth": [3, None],
              "clf__max_features": [1, 2,3],
              "clf__min_samples_split": [2, 3, 10],
              "clf__min_samples_leaf": [1, 3, 10],
              # "bootstrap": [True, False],
              "clf__criterion": ["gini", "entropy"]}
#pipeline.get_params().keys()
CV_rfc = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv= 10)
CV_rfc.fit(X_train, y_train)
print(CV_rfc.best_params_)
```
